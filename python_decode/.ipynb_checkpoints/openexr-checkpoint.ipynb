```python
import OpenEXR
import matplotlib.pyplot as plt
import numpy as np
import cv2

from skimage import restoration

%matplotlib inline
```

```python
# img_exr = OpenEXR.InputFile('/Users/albertxu/Desktop/view-0_1.exr')
img_exr = OpenEXR.InputFile('/Users/albertxu/CLionProjects/pbrt-v3-coded-aperture/killeroo-bare-depthtest.exr')
dw = img_exr.header()['dataWindow']
size = (dw.max.x - dw.min.x + 1, dw.max.y - dw.min.y + 1)
```

```python
cvim = cv2.imread('/Users/albertxu/CLionProjects/pbrt-v3-coded-aperture/killeroo-bare-depthtest.exr', -1)
cvim = np.clip(np.power(cvim, 0.45), 0, 1)
# cvim[cvim > 1] = 1
cvim *= 255
cvim = np.uint8(cvim)
cv2.imwrite('killeroo.png', cvim)
```

```python
img_exr.header()
```

```python
rch = img_exr.channel('R')
rimg = np.frombuffer(rch, dtype=np.float16)
rimg.shape = (size[1], size[0])
gch = img_exr.channel('G')
gimg = np.frombuffer(gch, dtype=np.float16)
gimg.shape = (size[1], size[0])
bch = img_exr.channel('B')
bimg = np.frombuffer(bch, dtype=np.float16)
bimg.shape = (size[1], size[0])
```

```python
ggim = np.stack([rimg, gimg, bimg], axis=-1)
# ggim *= 255 / np.amax(ggim)
# ggim[ggim > 255] = 255
ggim[ggim > .5] = .5
ggim *= 255*2
```

```python
plt.figure(figsize=(10,10))
plt.imshow(ggim.astype(np.uint8))
```

```python
rcpy = rimg.astype(np.float32)
rcpy[rcpy > 1] = 1
```

```python
plt.figure(figsize=(20,9))
plt.imshow(rcpy)
plt.colorbar()
```

```python
msk = cv2.imread('/Users/albertxu/CLionProjects/coded-masks/Masoudifar.png')
mskmin = cv2.resize(msk, (21, 21))[...,0] / 255
mskmin = mskmin / np.count_nonzero(mskmin)
```

```python
rd, _ = restoration.unsupervised_wiener(rcpy, mskmin)
# gd, _ = restoration.unsupervised_wiener(gimg, mskmin)
# bd, _ = restoration.unsupervised_wiener(bimg, mskmin)
```

```python
imd = np.stack([rd, gd, bd], axis=-1).astype(np.float32)
imd *= 255 / np.amax(imd)
imd[imd > 255] = 255
```

```python
plt.figure(figsize=(16,9))
plt.imshow(rd)
```

```python
plt.figure(figsize=(16,9))
plt.imshow(cv2.filter2D(rd.astype(np.float32), -1, mskmin))
plt.colorbar()
```
